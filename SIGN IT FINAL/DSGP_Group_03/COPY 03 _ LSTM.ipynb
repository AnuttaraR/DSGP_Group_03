```python
mp_holistic = mp.solutions.holistic # Holistic model
mp_drawing = mp.solutions.drawing_utils # Drawing utilities
```

```python
words = ['bye', 'hello', 'home', 'love', 'thankyou', 'angry','bank', 'brother', 'excuse me',
         'father', 'good evening', 'good morning', 'goodnight', 'happy', 'help', 'hospital',
         'how much', 'hungry', 'love', 'mother', 'police station', 'sad', 'school', 'sister',
         'sorry', 'welcome', 'what', 'when', 'where', 'who']
```

```python
words_2 = ['bye', 'hello', 'home', 'love', 'thankyou']
```

```python
import pickle

mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
dic = {}
with mp_hands.Hands(static_image_mode=False, max_num_hands=2, min_detection_confidence=0.5) as hands:
    for word in words:
        dire = os.path.join('src', word)
        vids = os.listdir(dire)
        vid_no = 0
        print(word)
        dic[word] = []
        for i,vid in enumerate(vids):
            dic[word].append([])
            vidcap = cv2.VideoCapture(os.path.join('src', word, vid))
            success,frame = vidcap.read()
            count = 0
            while success:     
                success,frame = vidcap.read()
                if success:
                    image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                    results = hands.process(image)
                    if results.multi_hand_landmarks:
                        # Loop through each hand landmark in the list
                        for hand_landmarks in results.multi_hand_landmarks:
                            # Get the landmarks for the hand
                            landmarks = hand_landmarks.landmark
#                             print(len(landmarks))
                            # Convert the landmarks to a numpy array
                            landmarks_array = np.array([[landmark.x, landmark.y, landmark.z] for landmark in landmarks])
                            # Store the landmarks in the dataset array
#                             dataset.append(np.array(landmarks).flatten())
                            dic[word][i].append(landmarks_array.flatten())
                            # Draw the hand landmarks on the image for visualization
                            mp_drawing.draw_landmarks(image, hand_landmarks, mp_hands.HAND_CONNECTIONS)
#                     npy_path = os.path.join('target_4', word, str(vid_no), str(count))
#                     np.save(npy_path, keypoints)
                
            vid_no +=1
        
```

```python
# create a binary pickle file 
f = open("preprocessed.pkl","wb")
pickle.dump(dic,f)
print("Successfuly transfered")
f.close()
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

src = {}
for key in data.keys():
    src[key] = []
    
for key in data.keys():
    arr = data[key]
    for vid in arr:
        if len(vid) < 35:
            src[key].append(vid + [[0] * 63] * (35 - len(vid)) )
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
```

```python
words = words
```

```python
label_map = {label:num for num, label in enumerate(words)}
```

```python
label_map
```

```python
src['bye'][0][0][0]
```

```python
sequences, labels = [], []
for word in words:
    for i,sequence in enumerate(src[word]):
        window = src[word][i][:35]
        sequences.append(np.asarray(window).astype('float32'))
        labels.append(label_map[word])
```

```python
np.array(sequences).shape
```

```python
np.array(labels).shape
```

```python
X = np.array(sequences)
```

```python
X.shape
```

```python
y = to_categorical(labels).astype(int)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
y_test.shape
```

# Build and Train LSTM Neural Network

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import TensorBoard
```

```python
log_dir = os.path.join('Logs')
tb_callback = TensorBoard(log_dir=log_dir)
```

```python
model = Sequential()
model.add(LSTM(64, return_sequences=True, activation='relu', input_shape=(35,63)))
model.add(LSTM(128, return_sequences=True, activation='relu'))
model.add(LSTM(64, return_sequences=False, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(30, activation='softmax'))
```

```python
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
model.fit(X_train, y_train, epochs=2000, callbacks=[tb_callback])
```
